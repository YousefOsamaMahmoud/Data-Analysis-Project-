```python
df = pd.read_csv("Untitled form (Responses) - Form Responses 1 (1).csv")
df.head(35)
```

```python
print("Column names:", df.columns)
```

```python
plt.figure(figsize=(8, 6))
df['Faculty'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Faculty')
plt.xlabel('Faculty')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['Academic Year'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Academic Year')
plt.xlabel('Academic Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
    plt.figure(figsize=(8, 6))
    df['How often do you find yourself procrastinating on academic tasks (e.g., Assignments, projects, studying for tests)?'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title('Frequency of Procrastination on Academic Tasks')
    plt.axis('equal')
    plt.show()
```

```python
    plt.figure(figsize=(8, 6))
    plt.hist(df['What is your CGPA'], bins=10, color='orange', edgecolor='black')
    plt.title('Distribution of CGPA')
    plt.xlabel('CGPA')
    plt.ylabel('Frequency')
    plt.show()
```

```python
    plt.figure(figsize=(8, 6))
    df['What are some common reasons or triggers for your procrastination?'].value_counts().plot(kind='bar', color='pink')
    plt.title('Common Reasons for Procrastination')
    plt.xlabel('Reasons')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
    plt.figure(figsize=(8, 6))
    df['Have you ever experienced negative consequences on your GPA as a result of procrastination?'].value_counts().plot(kind='bar', color='lightblue')
    plt.title('Negative Consequences on GPA due to Procrastination')
    plt.xlabel('Consequences')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
    plt.figure(figsize=(8, 6))
    plt.scatter(df['How often do you find yourself procrastinating on academic tasks (e.g., Assignments, projects, studying for tests)?'], df['What is your CGPA'])
    plt.title('Scatter Plot of Procrastination Frequency vs CGPA')
    plt.xlabel('Procrastination Frequency')
    plt.ylabel('CGPA')
    plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
    plt.grid(True)
    plt.show()
```
